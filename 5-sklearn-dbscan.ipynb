#### Load Dependencies

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
```

#### Set Hyperparameters

```python
PATH_DATA = "../data/"
```

#### Load Data

```python
iris = pd.read_csv(PATH_DATA + "iris.csv")
x = iris.drop(iris.species.name, axis=1)
x.head()
```

#### Train the Model

```python
db = DBSCAN(eps=0.5, min_samples=7)
db = db.fit(x)
clusters = db.labels_
clusters
```

#### Visualize the Results

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')  
fig, ax = plt.subplots(dpi=300, figsize=(8, 6))

scatter = ax.scatter(
    x.iloc[:, 0], 
    x.iloc[:, 2], 
    c=clusters,
    cmap='viridis',       
    s=60,               
    alpha=0.8,         
    edgecolor='white',   
    linewidth=0.5         
)

cbar = plt.colorbar(scatter)
cbar.set_label('Cluster ID', rotation=270, labelpad=15)

ax.set_xlabel("Sepal Length (cm)", fontsize=12, fontweight='bold')
ax.set_ylabel("Petal Length (cm)", fontsize=12, fontweight='bold') 
ax.set_title("Clustering Results", fontsize=14, fontweight='bold')

ax.grid(True, linestyle='--', alpha=0.3)

ax.set_xlim(x.iloc[:, 0].min()-0.5, x.iloc[:, 0].max()+0.5)
ax.set_ylim(x.iloc[:, 2].min()-0.5, x.iloc[:, 2].max()+0.5)

plt.tight_layout()
plt.show()
```
